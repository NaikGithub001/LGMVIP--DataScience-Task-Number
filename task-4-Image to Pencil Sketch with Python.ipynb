```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("E:/pics/1156126.jpg")
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.imshow(image)
plt.show()
cv2.imshow("lion", image)
cv2.waitKey(0)
```

# Converting an image into gray_scaler image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.imshow(gray_image)
plt.show()
cv2.imshow("New ", gray_image)
cv2.waitKey(0)
```

# Inverting the gray_image

```python
inverted_image = 255 - gray_image
plt.axis('off')
plt.imshow(inverted_image)
plt.show()
cv2.imshow("Inverted", inverted_image)
cv2.waitKey()
```

# Bluurring the inverted image

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
plt.axis('off')
plt.imshow(blurred)
plt.show()
```

# Inverting the blurred image

```python
inverted_blurred = 255 - blurred
plt.axis('off')
plt.imshow(inverted_blurred)
plt.show()
cv2.imshow("inverted", inverted_blurred)
cv2.waitKey(0)
```

# Pencil sketch image

```python
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
plt.axis('off')
plt.imshow(pencil_sketch)
plt.show()
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

```python
plt.axis('off')
plt.imshow(image)
plt.show()
plt.imshow(pencil_sketch)
plt.show()
cv2.imshow("original image", image)
cv2.imshow("pencil sketch", pencil_sketch)
cv2.waitKey(0)
```
